```python
import numpy
import math
from matplotlib import pyplot
```

```python
R = 1.15
r = numpy.linspace(R,5,100)
theta = numpy.linspace(0,2*numpy.pi,145)
```

```python
radius_matrix, theta_matrix = numpy.meshgrid(r,theta)
```

```python
def get_velocity_doublet(strength, r,theta):
    """Returns the velocity field generated by a doublet.
    
    Arguments
    ---------
    strength -- strength of the doublet.
    xd, yd -- coordinates of the doublet.
    X, Y -- mesh grid.
    """
    v_r = - strength/(2*numpy.pi)*numpy.cos(theta)/r**2
    v_t = - strength/(2*numpy.pi)*numpy.cos(theta)/r**2
    
    return v_r, v_t
```
