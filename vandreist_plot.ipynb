```python
import math
import numpy
from matplotlib import pyplot
from scipy import integrate
```

```python
y = numpy.linspace(1,100,1001)
nx=10
```

```python
def dreistvelocity(k):
    def func(p):
        return (2/(1+(1+4*0.4*0.4*p**2*(1-numpy.exp(-p/25))**2)**0.5))
    result = 0
    result = integrate.quad(lambda p: func(p), 0, k)[0]
    return result
```

```python
def prandtlvelocity(k):
    z=0
    if k>11.6:
        z=2.5*numpy.log(k)+5.5
    else:
        z=k
    
    return z
```

```python
u = numpy.zeros(len(y))
v = numpy.zeros(len(y))
for i in range(len(y)):
    u[i]=dreistvelocity(y[i])
    v[i]=prandtlvelocity(y[i])
```

```python
%matplotlib inline
pyplot.figure(figsize=(12, 8))
pyplot.xlim(1,100)
pyplot.grid(True)
pyplot.plot(y,u,color='b',label='Van Driest A=25 k=0.4')
pyplot.plot(y,v,color='r',label='Prandtl and Taylor')
pyplot.xscale('log')
pyplot.legend(loc='best', prop={'size':14})
pyplot.title('Velocity Measurement in Turbulent Flow', fontsize=16)
```
