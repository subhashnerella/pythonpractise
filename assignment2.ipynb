```python
import numpy
import math
from matplotlib import pyplot
```

```python
R = 1.15
r = numpy.linspace(R,5,100)
theta = numpy.linspace(0,2*numpy.pi,145)
```

```python
radius_matrix, theta_matrix = numpy.meshgrid(r,theta)
X = (radius_matrix * numpy.cos(theta_matrix)-0.15)
Y = radius_matrix * numpy.sin(theta_matrix)
```

```python
def get_joukouskipoints(z):
    q =z+1/z
    return q
```

```python
z=numpy.vectorize(complex)(X,Y)
w=get_joukouskipoints(z)
```

```python
%matplotlib inline
pyplot.figure(figsize=(5,5))
pyplot.scatter(X,Y,color='#CD2305', s=0.125, marker='o')
```

```python
%matplotlib inline
pyplot.figure(figsize=(5,5))
pyplot.xlim(-6,6)
pyplot.ylim(-6,6)
pyplot.scatter(w.real,w.imag,color='#CD2305', s=0.125, marker='o')
```

```python
kappa = 8.309                        # strength of the doublet
x_doublet, y_doublet = -0.15, 0.0 
```

```python
def get_velocity_doublet(strength, xd, yd, X, Y):
    """Returns the velocity field generated by a doublet.
    
    Arguments
    ---------
    strength -- strength of the doublet.
    xd, yd -- coordinates of the doublet.
    X, Y -- mesh grid.
    """
    u = - strength/(2*math.pi)*((X-xd)**2-(Y-yd)**2)/((X-xd)**2+(Y-yd)**2)**2
    v = - strength/(2*math.pi)*2*(X-xd)*(Y-yd)/((X-xd)**2+(Y-yd)**2)**2
    
    return u, v
```

```python
u_doublet, v_doublet = get_velocity_doublet(kappa, x_doublet, y_doublet, X, Y)
```

```python
%matplotlib inline

# plots the streamlines
size = 10
pyplot.figure(figsize=(size,size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(-6, 6)
pyplot.ylim(-6, 6)
pyplot.streamplot(X, Y, u_doublet,v_doublet,density=0.5, linewidth=0.5, arrowsize=0.5, arrowstyle='->')
pyplot.scatter(x_doublet, y_doublet, color='#CD2305', s=80, marker='o');
```
